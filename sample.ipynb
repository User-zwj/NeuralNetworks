# Normal Samples

```python
import numpy as np
from scipy.stats import norm

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-3,3,100)
y1 = np.random.normal(0,1,size=100)

plt.figure()
plt.plot(x,y1,label='Normal random samples')
plt.legend();
```

```python
plt.hist(y1);
```

```python
plt.figure()
y2 = norm.pdf(x, loc=0, scale=1)
plt.plot(x,y2,'r',label='Normal density')
plt.legend();
```
